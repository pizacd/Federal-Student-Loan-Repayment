```python
#studentaid.gov numbers until 07/2020
#undergraduate rate is 4.53%, graduate rate is 6.08%
import matplotlib.pyplot as plt
def fed_loans(amount, payment, rate = .0453):
    
    def month_int(balance, rate):
        return round(rate/12*balance,2)
    # number from salliemaecalculator

    #start with avg loan
    #calculate one principal payment
    #reduce principal payment
    loan_plot = [amount]
    balance = amount
    while balance>0:
        balance = round(balance-(payment-month_int(balance,rate)),2)
        loan_plot.append(balance)
        continue

    #start 0% int scenario by calculating 6 months worth of no interest payments
    no_int = []
    for i in range(0,7):
        no_int.append(amount-payment*i)    
    no_int_balance = no_int[-1] #starting balance when interest rate kicks back in
    while no_int_balance>0:
        no_int_balance = round(no_int_balance-(payment-month_int(no_int_balance,rate)),2)
        no_int.append(no_int_balance)
        continue   

    no_pay = loan_plot.copy() # taking loan plot and adding starting balance to list 6x to simulate no payments made
    for i in range(0,6):
        no_pay.insert(i,amount)

    plt.figure(figsize = (20,15))
    plt.grid(linestyle='-', linewidth=2)
    plt.plot(loan_plot,'o',label = 'Normal Student Loan Payment Plan')
    plt.plot(no_int,'o', label = 'Paying 0% Interest Until 9/30/20')
    plt.plot(no_pay,'o', label = 'Deferring Student Loan Payments Until After 9/30/20')
    plt.xlabel('Months Paying Federal Student Loans',fontsize = 15)
    plt.ylabel('Remaining Federal Student Loan Balance (US Dollars)',fontsize = 15)
    plt.title('Paying Federal Student Loans During COVID-19 Relief',fontsize = 25,fontweight = 'bold')
    plt.legend(loc = 'upper right',fontsize = 12)
    print('Paying your federal student loans during the COVID-19 relief period means you will make {} fewer payments\n than under the normal payment system.\n This will save you an estimated ${}.'.format(len(loan_plot)-len(no_int),payment*(len(loan_plot)-len(no_int))))
```
